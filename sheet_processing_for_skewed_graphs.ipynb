```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./pokemon_data.csv")
print(df["Defense"].skew())
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sb.distplot(df["Defense"])
plt.subplot(1,2,2)
sb.boxplot(y="Defense", data=df)
```

```python
# we need 25% percentile and 75% 
# df["Defense"].describe()

percentile25 = df["Defense"].quantile(0.25)
percentile75 = df["Defense"].quantile(0.75)
print(percentile25, percentile75)

iqr = percentile75 - percentile25
print(iqr)

h_bound = percentile75 + 1.5 * iqr
l_bound = percentile25 - 1.5 * iqr

print(h_bound, l_bound)
```
