```python
# formate the born and died date
year_pattern = r"(\d{4})"

df['Born_Year'] = pd.to_numeric(df['Born'].str.extract(year_pattern)[0],  errors='coerce').astype('Int64')

df['Died_Year'] = pd.to_numeric(df['Died'].str.extract(year_pattern)[0],  errors='coerce').astype('Int64')
df.head()

```
